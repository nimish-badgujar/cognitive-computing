```

# Task 1: List Operations
print("Task 1: List Operations")
L = [10, 20, 30, 40, 50, 60, 70, 80]
print("Initial List:", L)

# i. Add 200 and 300 to L
L.append(200)
L.append(300)
print("After adding 200 and 300:", L)

# ii. Remove 10 and 30 from L
L.remove(10)
L.remove(30)
print("After removing 10 and 30:", L)

# iii. Sort L in ascending order
L.sort()
print("Ascending order:", L)

# iv. Sort L in descending order
L.sort(reverse=True)
print("Descending order:", L)
```

```

# Task 2: Tuple Operations
print("Task 2: Tuple Operations")
scores = (45, 89.5, 76, 45.4, 89, 92, 58, 45)
print("Scores Tuple:", scores)

# i. Highest score and its index
highest_score = max(scores)
highest_index = scores.index(highest_score)
print("Highest Score:", highest_score, "at index", highest_index)

# ii. Lowest score and its count
lowest_score = min(scores)
lowest_count = scores.count(lowest_score)
print("Lowest Score:", lowest_score, "appears", lowest_count, "times")

# iii. Reverse the tuple and return as a list
reversed_list = list(reversed(scores))
print("Reversed Tuple as List:", reversed_list)

# iv. Check if a specific score (76) is present
user_score = 76  # Change this value for input
if user_score in scores:
    print(f"Score {user_score} is present at index {scores.index(user_score)}")
else:
    print(f"Score {user_score} is not present in the tuple")
```

```

# Task 3: Random Number List
import random

print("Task 3: Random Number List")
random_numbers = [random.randint(100, 900) for _ in range(100)]
print("Generated Numbers:", random_numbers)

# i. Odd numbers
odd_numbers = [num for num in random_numbers if num % 2 != 0]
print("Odd Numbers:", odd_numbers)

# ii. Even numbers
even_numbers = [num for num in random_numbers if num % 2 == 0]
print("Even Numbers:", even_numbers)

# iii. Prime numbers
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

prime_numbers = [num for num in random_numbers if is_prime(num)]
print("Prime Numbers:", prime_numbers)
```

```

# Task 4: Set Operations
print("Task 4: Set Operations")
A = {34, 56, 78, 90}
B = {78, 45, 90, 23}
print("Set A:", A)
print("Set B:", B)

# i. Unique scores (union)
unique_scores = A.union(B)
print("Union of A and B:", unique_scores)

# ii. Common scores (intersection)
common_scores = A.intersection(B)
print("Intersection of A and B:", common_scores)

# iii. Exclusive scores (symmetric difference)
exclusive_scores = A.symmetric_difference(B)
print("Symmetric Difference of A and B:", exclusive_scores)

# iv. Subset and superset check
print("Is A a subset of B?", A.issubset(B))
print("Is B a superset of A?", B.issuperset(A))

# v. Remove a specific score from A
X = 78  # Change this value for user input
if X in A:
    A.remove(X)
    print(f"Score {X} removed from A. Updated Set A:", A)
else:
    print(f"Score {X} is not present in Set A")
```

```

# Task 5: Dictionary Operations
print("Task 5: Dictionary Operations")
location_dict = {"city": "New York", "country": "USA", "population": "8M"}
print("Original Dictionary:", location_dict)

# Rename key "city" to "location"
if "city" in location_dict:
    location_dict["location"] = location_dict.pop("city")
    print("Updated Dictionary:", location_dict)
else:
    print("'city' key not found in the dictionary")
```
