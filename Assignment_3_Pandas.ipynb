## Q6: Create an employee dataset

```python

# Creating employees dataset
employees = pd.DataFrame({
    'Employee_ID': [101, 102, 103, 104, 105],
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Edward'],
    'Department': ['HR', 'IT', 'IT', 'Marketing', 'Sales'],
    'Age': [29, 34, 41, 28, 38],
    'Salary': [50000, 70000, 65000, 55000, 60000],
    'Years_of_Experience': [4, 8, 10, 3, 12],
    'Joining_Date': ['2020-03-15', '2017-07-19', '2013-06-01', '2021-02-10', '2010-11-25'],
    'Gender': ['Female', 'Male', 'Male', 'Female', 'Male'],
    'Bonus': [5000, 7000, 6000, 4500, 5000],
    'Rating': [4.5, 4.0, 3.8, 4.7, 3.5]
})

employees
```

## Q6: Perform various DataFrame operations

```python

# a) Shape of DataFrame
shape = employees.shape

# b) Summary of DataFrame
summary = employees.info()

# c) Descriptive statistics
stats = employees.describe()

# d) First 5 and last 3 rows
first_five = employees.head()
last_three = employees.tail(3)

# e) Various statistics
avg_salary = employees['Salary'].mean()
total_bonus = employees['Bonus'].sum()
youngest_age = employees['Age'].min()
highest_rating = employees['Rating'].max()

# f) Sort by salary
sorted_by_salary = employees.sort_values(by='Salary', ascending=False)

# g) Categorize performance ratings
def categorize_performance(rating):
    if rating >= 4.5:
        return 'Excellent'
    elif rating >= 4.0:
        return 'Good'
    else:
        return 'Average'

employees['Performance_Category'] = employees['Rating'].apply(categorize_performance)

# h) Identify missing values
missing_values = employees.isnull().sum()

# i) Rename Employee_ID column
employees.rename(columns={'Employee_ID': 'ID'}, inplace=True)

# j) Filter employees
experienced_employees = employees[employees['Years_of_Experience'] > 5]
it_employees = employees[employees['Department'] == 'IT']

# k) Add Tax column
employees['Tax'] = employees['Salary'] * 0.10

# l) Save modified DataFrame
employees.to_csv('modified_employees.csv', index=False)
```
